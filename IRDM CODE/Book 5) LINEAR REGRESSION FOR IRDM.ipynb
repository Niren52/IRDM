# FUNCTIONS

```
#CSV TO LIST
def load_csv(filename):
  dataset = list()
  with open(filename, 'r') as file:
    csv_reader = reader(file)
    for row in csv_reader:
      if not row:
        continue
      dataset.append(row)
  del dataset[0]    
  return dataset
```

```
#STRINGS TO NUMERIC VALUES
def str_column_to_float(dataset, column):
	for row in dataset:
		row[column] = float(row[column].strip())
```

```
#FUNCTION TO CREATE ROOT MEAN SQUARED ERROR:
def rmse_metric(actual, predicted):
	sum_error = 0.0
	for i in range(len(actual)):
		prediction_error = predicted[i] - actual[i]
		sum_error += (prediction_error ** 2)
	mean_error = sum_error / float(len(actual))
	return np.sqrt(mean_error)
```

```
##MODEL ACCURACY AND ERROR METRICS
def EVALUATION(dataset,dataset2, linearRegression, l_rate, n_epoch):
  scores = list()
  train_set = dataset
  test_set = dataset2
  predicted = linearRegression(train_set, test_set, l_rate, n_epoch)
  actual = [row[-1] for row in test_set]
  rmse = rmse_metric(actual, predicted)
  scores.append(rmse)
  print("RMSE: " + str(scores))
  return
```

```
#Make prediction with optimised coeffecients
def predict(row, coefficients):
	yhat = coefficients[0]
	for i in range(len(row)-1):
		yhat += coefficients[i + 1] * row[i]
	return yhat
```

```
# #STOCHASTIC GRADIENT COEFFECIENT ESTIMATION
def coefficients_sgd(train, l_rate, n_epoch):
	coef = [0.0 for i in range(len(train[0]))]
	for epoch in range(n_epoch):
		for row in train:
			yhat = predict(row, coef)
			error = yhat - row[-1]
			coef[0] = coef[0] - l_rate * error
			for i in range(len(row)-1):
				coef[i + 1] = coef[i + 1] - l_rate * error * row[i]
			# print(l_rate, n_epoch, error)
	return coef
```

```
#LINEAR REGRESSION USING STOCHASTIC GRADIENT DESCENT
def linear_regression_sgd(train, test, l_rate, n_epoch):
	predictions = list()
	coef = coefficients_sgd(train, l_rate, n_epoch)
	for row in test:
		yhat = predict(row, coef)
		predictions.append(yhat)
	return(predictions)
```

```
#FUNCTION FOR HYPERPAPRAMETER TUNING
def TrainingForHyperTuning(l_rate, n_epoch, filename,filename2):
  seed(123) # For reproducibility
  # load and prepare data
  dataset = load_csv(filename)
  dataset2 = load_csv(filename2)
  for i in range(len(dataset[0])):
    str_column_to_float(dataset, i)
  for i in range(len(dataset2[0])):
    str_column_to_float(dataset2, i)
  scores = EVALUATION(dataset, dataset2, linear_regression_sgd, l_rate, n_epoch)
  return
```
