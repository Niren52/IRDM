# FUNCTIONS

```
#Read file to List Formats
def load_csv(filename):
  dataset = list()
  with open(filename, 'r') as file:
    csv_reader = reader(file)
    for row in csv_reader:
      if not row:
        continue
      dataset.append(row)
  del dataset[0]    
  return dataset
```

```
#STRINGS TO NUMERIC VALUES
def str_column_to_float(dataset, column):
	for row in dataset:
		row[column] = float(row[column].strip())
```

```
#PREDICT CORRECT AND INCORRECT VALUES
def accuracy_metric(actual, predicted):
	correct = 0
	for i in range(len(actual)):
		if actual[i] == predicted[i]:
			correct += 1
	return correct / float(len(actual)) * 100.0
```

```
#FUNCTION TO EVALUATE THE EFFECTIVENESS OF LOGISTIC REGRESSION
def evaluate_algorithm(dataset, algorithm, dataset2, *args):
  scores = list()
  train_set = dataset
  predicted = algorithm(train_set, dataset2, *args)
  actual = [row[-1] for row in dataset2]
  accuracy = accuracy_metric(actual, predicted)
  scores.append(accuracy)
  return scores
 
#MAKE PREDICTIONS WITH UPDATED COEFFICIENTS
def predict(row, coefficients):
	yhat = coefficients[0]
	for i in range(len(row)-1):
		yhat += coefficients[i + 1] * row[i]
	return 1.0 / (1.0 + exp(-yhat))
 
#STOCHASTIC GRADIENT COEFFECIENT ESTIMATION
def coefficients_sgd(train, l_rate, n_epoch):
  coef = [0.0 for i in range(len(train[0]))]
  for epoch in range(n_epoch):
    for row in train:
      yhat = predict(row, coef)
      error = row[-1] - yhat
      coef[0] = coef[0] + l_rate * error * yhat * (1.0 - yhat)
      for i in range(len(row)-1):
        coef[i + 1] = coef[i + 1] + l_rate * error * yhat * (1.0 - yhat) * row[i]
  print("MSE:   " + str(error))
  return coef
 
#LOGISTIC REGRESSION USING STOCHASTIC GRADIENT DESCENT
def logistic_regression(train, test, l_rate, n_epoch):
	predictions = list()
	coef = coefficients_sgd(train, l_rate, n_epoch)
	for row in test:
		yhat = predict(row, coef)
		yhat = round(yhat)
		predictions.append(yhat)
	return(predictions)
```

```
#ACCURACY AND ERROR METRIC REPORTING
def Evaluation(l_rate,n_epoch,filename,filename2):
  seed(123) # For reproducibility
  # load and prepare data
  #filename = 'TrainDiscuss.csv'
  #filename2 = 'ValidationDiscuss.csv'
  dataset = load_csv(filename)
  dataset2 = load_csv(filename2)
  for i in range(len(dataset[0])):
    str_column_to_float(dataset, i)
  for i in range(len(dataset2[0])):
    str_column_to_float(dataset2, i)
  scores = evaluate_algorithm(dataset, logistic_regression, dataset2, l_rate, n_epoch)
  print('Accuracy: %.3f%%' % (sum(scores)/float(len(scores))))
  return
```

#TESTING: USING BEST RESULTS FROM HYPERPARAMETER TUNING

```
#DEFINE ALGORITHM TO GET PREDICTIONS BETWEEN 0 and 1
def EVALUATE_TEST(dataset, algorithm, dataset2, *args):
  train_set = dataset
  predicted = algorithm(train_set, dataset2, *args)
  return predicted
 
# Estimate logistic regression coefficients using stochastic gradient descent
def coefficients_sgd(train, l_rate, n_epoch):
  coef = [0.0 for i in range(len(train[0]))]
  for epoch in range(n_epoch):
    for row in train:
      yhat = predict(row, coef)
      error = row[-1] - yhat
      coef[0] = coef[0] + l_rate * error * yhat * (1.0 - yhat)
      for i in range(len(row)-1):
        coef[i + 1] = coef[i + 1] + l_rate * error * yhat * (1.0 - yhat) * row[i]
  return coef
 
# Linear Regression Algorithm With Stochastic Gradient Descent
def logistic_regression(train, test, l_rate, n_epoch):
	predictions = list()
	coef = coefficients_sgd(train, l_rate, n_epoch)
	for row in test:
		yhat = predict(row, coef)
		predictions.append(yhat)
	return(predictions)

# Make a prediction with coefficients
def predict(row, coefficients):
	yhat = coefficients[0]
	for i in range(len(row)-1):
		yhat += coefficients[i + 1] * row[i]
	return 1.0 / (1.0 + exp(-yhat))
```
